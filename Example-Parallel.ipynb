```python
def tpsmall(lab, q, a, l = 1e-2):
    '''Print where the function (a) exceed the limit l'''
    mid = np.where(q >1)[0][0]
    mid2 = np.where(q >100)[0][0]
    try:
        q1 = q[np.where(a[:mid] <l)[0][-1]]
    except:
        q1 = 'Big'
    try:
        q2 = q[mid2 + np.where(a[mid2:] <l)[0][0]]
    except:
        q2 = 'Big'
    print(lab, q1, q2)
```
